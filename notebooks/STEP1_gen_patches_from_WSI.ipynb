```python
import os
from utils.utils import interpolate, generate_patches_from_wsi

vis_scale = 0.007
enlarge = 5
perc_wpx = 0.70
perc_bpx = 0.05
real_enlarge = int(enlarge / vis_scale)
patches_path = "../data/patches/"
coords_path = "../checkpoints/coords_checkpoints"
overview_path = "../results/overview_wsis"
```

## Paul-Brousse

```python
step_pb = 1152
patch_size_pb = 1152
path_to_PB_wsis = "../data/WSIs/PB"
slides = [
    e + "/" + f
    for e in os.listdir(f"{path_to_PB_wsis}")
    for f in os.listdir(f"{path_to_PB_wsis}/{e}")
    if ".mrxs" in f
]
slides.sort()
print(slides)
os.makedirs('../data/patches/',exist_ok=True)
os.makedirs('../results/overview_wsis/',exist_ok=True)
os.makedirs('../checkpoints/coords_checkpoints',exist_ok=True)
```

```python
for slide_name in slides:
    print(slide_name, "...")
    if slide_name.split("/")[-1].split(".")[0] not in os.listdir(patches_path):
        generate_patches_from_wsi(
            slide_name,
            path_to_wsi=path_to_PB_wsis,
            patch_size=patch_size_pb,
            step=step_pb,
            path_to_patches=patches_path,
            vis_scale=vis_scale,
            overview_path=overview_path,
            coords_path=coords_path,
            perc_wpx=perc_wpx,
            perc_bpx=perc_bpx,
            enlarge=enlarge,
        )
    else:
        print(slide_name, "exists")
```

## Henri-Mondor

```python
patch_size_hm = interpolate(patch_size_pb, 0.25, 0.46)
print(patch_size_hm)
step_hm = patch_size_hm

path_to_HM_wsis = "../data/WSIs/HM"
os.makedirs(path_to_HM_wsis,exist_ok=True)
slides = [
    e + "/" + f
    for e in os.listdir(f"{path_to_HM_wsis}")
    for f in os.listdir(f"{path_to_HM_wsis}/{e}")
    if ".ndpi" in f and "ndpa" not in f
]
slides.sort()
slides.reverse()
print(slides)
```

```python
for slide_name in slides:
    print(slide_name, "...")
    if slide_name.split("/")[-1].split(".")[0] not in os.listdir(patches_path):
        generate_patches_from_wsi(
            slide_name,
            path_to_wsi=path_to_HM_wsis,
            patch_size=patch_size_hm,
            step=step_hm,
            path_to_patches=patches_path,
            vis_scale=vis_scale,
            overview_path=overview_path,
            coords_path=coords_path,
            perc_wpx=perc_wpx,
            perc_bpx=perc_bpx,
            enlarge=enlarge,
        )
    else:
        print(slide_name, "exists")
```
